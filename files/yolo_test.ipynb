```python
# PyTorch Hub
import torch

model = torch.hub.load('ultralytics/yolov5', 'custom', path='drive/MyDrive/best.pt')  # default
```

```python
import os.path
 
targerdir = "drive/MyDrive/class"
 
file_list = os.listdir(targerdir)
img_list = []

for f in file_list:
    img_list.append(targerdir + "/" + f)

print(len(img_list))
```

```python
import cv2
import numpy as np
#from google.colab.patches import cv2_imshow
import easyocr

for i in img_list:
  path=i
 
  results = model(path)
  img = cv2.imread(path)

  if type(img) == type(None):
    continue
  # if type(img) == type(None):
  #   continue;
  # print(type(img))

  #cv2_imshow(img)
  #results.show()

  roi=results.xyxy[0][:, :-2].cpu().data.numpy()
  roi=roi.astype(np.int32)
  class_name = []

  s=path.split('/')
# s[-1]
  f_path='drive/MyDrive/directory/'+s[-1]
  os.makedirs(f_path)

  for i in range(len(roi)):
    img2=img[roi[i][1]:roi[i][3],roi[i][0]:roi[i][2]]
    #cv2_imshow(img2)
    #OCRÌïòÎäî ÌîÑÎ°úÍ∑∏Îû® Ïã§Ìñâ
    print("******************OCRÏûëÎèô~~~!@!@**************")
    cv2.imwrite('new.png', img2)

    path='new.png'
  
    reader = easyocr.Reader(['en'], gpu=True) # need to run only once to load model into memory
    result = reader.readtext(path)
    # print(type(result))
    # print(len(result))

    new_chars = []
    org_chars= []
    result_chars = ''
    # ban_chars=['interface','abstract']

    for i in range(len(result)):
      text=result[i][1]
      org_chars.append(result[i][1]) 
      for c in text:
        if (ord('Í∞Ä') <= ord(c) <= ord('Ìû£')) or (ord('A') <= ord(c) <= ord('Z')) or (ord('a') <= ord(c) <= ord('z') or (ord('(') <= ord(c)<=ord(')')) or  ord('<') == ord(c) or ord('>') == ord(c) ):
          result_chars += c
        else:
          result_chars += ''
      if(len(result_chars)>0):
        if(result_chars[0]=='('or result_chars[0]=='<' or result_chars[-1]=='>' or result_chars[-1]==')'): 
          result_chars = ''
          continue

      new_chars.append(result_chars)  
      result_chars = ''
      break
    if(len(new_chars)>0):
      print(new_chars[0]+"\n\n")
      class_name.append(new_chars[0])
      cv2.imwrite(f_path+'/'+new_chars[0]+'.jpg', img2) 
    # print("++++++++++++++++++++++++++++++++++")
    # print(new_chars)
  # #print(org_chars)
  # print(type(new_chars))
  # print(new_chars[0])
  



print(class_name )
  
# cv2_imshow(img)
# cv2.imwrite('Í≤ΩÎ°ú1', ima)
```
